```python
import pandas as pd
import numpy as np
# xl = "training_data_emo.xlsx"
# xl = pd.ExcelFile(r'training_data_emo.xlsx')
# # df = pd.read_excel(file_loc, index_col=None, na_values=['NA'], usecols = "A:B")
# ncols = xl.book.sheet_by_index(0).ncols
# df = xl.parse(0, converters={i : str for i in range(ncols)})
df = pd.read_excel('training_data_actual.xlsx', sheet_name=0) # can also index sheet by name or fetch all sheets
sent_list = df['label'].tolist()
# print(sent_list)

tweet_list = df['tweet'].tolist()
# print(tweet_list)

print(len(sent_list))
print(len(tweet_list))
```

```python
#tweet list
words = []
for tweet in tweet_list:
    sent = tweet.split()
    words.extend(sent)
# print(words)

# print(words)
```

```python
#tweet_list & sent_list
word_lists = []
tweet_lists= []
for tweet in tweet_list:
    word_lists = tweet.split()
    tweet_lists.append(word_lists)
# print(tweet_lists)
    
    
tweet_tuple =  tuple(zip(tweet_lists, sent_list))
print(tweet_tuple)
```

```python
# run this, inspect the output, make sure you understand the output
all_word_freqs = nltk.FreqDist(w.lower() for w in words)
print(all_word_freqs)
# for w in words:
#     print(w)
```

```python
# this function takes a document (as a string) and returns the
# feature values (word counts) for all the words in the training data
document = 'predictable no fun'.split()
def document_features(document):
    document_words = set(document)
    features = {}
    for word in all_word_freqs:
        features['contains({})'.format(word)] = (word in document_words)
    return features

print(document_features(document))
```

```python
# if you have completed the steps above, this code cell should output the 
# same class that you computed in Part 1
train_set = [(document_features(d), c) for (d,c) in tweet_tuple]
classifier = nltk.NaiveBayesClassifier.train(train_set)
classifier.classify(document_features(''.split()))
```
